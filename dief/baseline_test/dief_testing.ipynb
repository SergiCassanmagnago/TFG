# ca-AstroPh.csv

```python
#WRITING THE METRICS CSV FILE FOR ca-AstroPh:

# csv header
fieldnames = ['test','approach','tfft','totaltime','comp']

# csv data
rows = [
    ['ca-AstroPh', 'DP-WCC', 0.04418945312, 19.46020508, 290],
    ['ca-AstroPh', 'Haskell-WCC', 25.713135, 25.713135, 290]
]
with open('/content/drive/MyDrive/TFG/dief/ca-AstroPh-metrics.csv', 'w', encoding='UTF8', newline = "") as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(fieldnames)

    # write multiple rows
    writer.writerows(rows)
```

```python
df = pd.read_csv('/content/drive/MyDrive/TFG/dief/ca-AstroPh-metrics.csv')
df.head()
```

```python
COLORS = ["#ECC30B", "#D56062"]

# Load the result of the metrics
metrics_astro = diefpy.load_metrics("/content/drive/MyDrive/TFG/dief/ca-AstroPh-metrics.csv")

# Plot the execution times of all queries
diefpy.plot_execution_time(metrics_astro, COLORS, log_scale=True).show()
```

```python
# Load the answer trace file with the query traces
traces_astro = diefpy.load_trace("/content/drive/MyDrive/TFG/dief/ca-AstroPh.csv")

# Plot the answer trace
diefpy.plot_answer_trace(traces_astro, "ca-AstroPh", COLORS).show()
```

```python
dt = diefpy.dieft(traces_astro, "ca-AstroPh", 10)
pd.DataFrame(dt).head()
```

```python
exp1 = diefpy.performance_of_approaches_with_dieft(traces_astro, metrics_astro)
diefpy.plot_performance_of_approaches_with_dieft(exp1, "ca-AstroPh", COLORS).show()
```
